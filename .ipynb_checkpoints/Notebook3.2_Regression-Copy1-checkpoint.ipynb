# Predicción numérica mediante regresión lineal

Este notebook cubre algunas de las formas básicas de realizar análisis de regresión en un conjunto de datos y predecir una variable continua.

## Linear Regression  

```python
# Librerías a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
from scipy import stats
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")

from sklearn.linear_model import LinearRegression #Regresión Lineal con scikit-learn
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from sklearn import preprocessing # para el escalado

from sklearn.feature_selection import SelectFromModel # para selección de características
```

## About the dataset: Boston housing

<b>Visualizar</b> para ver distribución de cada variable, determinar si hay o no valores atípicos y determinar si hay relación lineal entre variables independientes con y.

```python
def plot_base_relation(df, y, figsize=(20, 60)):
    columns = df.columns.tolist()
    _, axs = plt.subplots(len(columns), 3, figsize=figsize)
    
    for idx, column in enumerate(columns):
        # To get distribution of data
        sns.histplot(
            x=df[column],
            kde=False,
            color='#65b87b', alpha=.7,
            ax=axs[idx][0]
        )

        # To get knowledge about outliers
        sns.boxplot(
            x=df[column],
            color='#6fb9bd',
            ax=axs[idx][1]
        )

        # To get its relation with y
        sns.scatterplot(
            x=df[column], y=y, 
            color='#706dbd',
            ax=axs[idx][2]
        )
```

```python
plot_base_relation(boston[x_columns], boston[y_column], figsize=(12, 42))
```

De lo que se observa, no hay relaciones lineales significativas entre las variables independientes (X) y la variable (y). Aunque se aprecia que la variable con la que mayor relación positiva hay es RM y con la relación inversa más fuerte es con LSTAT.

Vamos a obtener los coeficientes de correlación para comprobarlo.

```python
correlation = boston.corr(); correlation['Reason for absence']
```

La comprobación anterior nos ayuda a anticipar si podríamos tener éxito con el modelo de regresión lineal, en este caso observamos que, hay dos variables con una relación mayor, aunque los coeficientes no son significativos. En todo caso, vamos a llevar a cabo diferentes experimentos, con diferentes versiones del dataset, para ver cómo se comporta la regresión lineal.

Antes de llevar a cabo los experimentos, para conseguir mejores resultados, mediante regresión lineal, es importante:

- Verificar valores nulos, que según lo comprobamos no existen.
- Realizar el tratamiento de valores atípicos. En la visualización previa, detectamos que hay 4 variables con outliers: CRIM, ZN, B y RM. Por las primeras 3 variables no nos vamos a preocupar porque no tienen una relación lineal significativa con MEDV, e incluso las podemos eliminar. En el caso de RM, vamos a realizar dos experimentos, uno con tratamiento de outliers y otro sin para ver si hay algún efecto positivo.
- Transformar variables categóricas al Onehot encoding.
- Realizar feature engineering. Para regresión lineal múltiple o para aplicar métodos de regresión donde hay múltiples variables predictoras, se recomienda aplicar este proceso. Más adelante vamos a ver algunas opciones.

## 1. Predicción mediante Regresión Lineal Simple

### 1.1 Experimento con las variables con más alta correlación (+ y -)

#### Correlación más altas.

- correlación(MEDV, RM) = 0.695360
- correlación(MEDV, LSTAT) = -0.737663

#### 1.1.1 Experimento 1: RLS usando RM como x (sin tratamiento de outliers)

```python
### Preparación de datasets (entrenamiento y test)
from sklearn.model_selection import train_test_split

x = boston['LSTAT']
y = boston['MEDV']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle = True, random_state=10)
```

```python
x_train.shape # 372
x_test.shape # 102
```

```python
############################################
#PREDICCIÓN: crear el modelo predictivo.
############################################

rl1 = LinearRegression() # creamos una instancia de LinearRegression

# Instruimos al modelo que aprenda (fit) de los datos (x,y)
rl1.fit(x_train.index.values.reshape(-1, 1), y_train)  # x.reshape(-1,1): redimensiona x en arrays de 1.

# Parámetros que ha estimado la regresión lineal:
print('a = ' + str(rl1.intercept_) , 'y  b = ' + str(rl1.coef_) )
```

```python
###################
#Utilizando el modelo de regresión para predecir nuevos valores (usar el método predict)
###################

# Utilizar el modelo para predecir con el conjunto de training:
y_train_predicted = rl1.predict(x_train.index.values.reshape(-1,1))

# Utilizar el modelo para predecir con el conjunto de test:
y_test_predicted = rl1.predict(x_test.index.values.reshape(-1,1))

#Ver valores reales vs predichos:
lr_predictions = pd.DataFrame({'Actual':y_test,'Predicted':y_test_predicted})
lr_predictions[:10]
```

```python
###################
#Visualización de los resultados (Para saber cómo se está comportando el modelo)
###################

# Visualizar los valores reales de y (líneas azuales) vs. los valores predichos (puntos negros).

pd.Series(y_test[:30].tolist()).plot(color='steelblue')  # Valores reales
pd.Series(y_test_predicted[:30].tolist()).plot(marker="o", linewidth=0, color='black') # Valores generados por el modelo
```

Como se observa en el último dataframe y en la gráfica anterior, hay muy pocos aciertos en la predicción, y los errores más grandes se produce en los valores atípicos de y (MEDV).

```python
###################
### Evaluación mediante métricas:
###################

# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = y_test_predicted)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))

# Coeficiente de determinación R^2 (usamos el método score).

# r^2 sobre el dataset de entrenamiento:
print('Coeficiente de Determinación R2 (train) = ' + str(r2_score(y_train, y_train_predicted)))

# r^2 sobre el dataset de test:
print('Coeficiente de Determinación R2 (test) = ' + str(r2_score(y_test, y_test_predicted)))
```

Como se observa en el último dataframe y en la gráfica anterior, hay muy pocos aciertos en la predicción, y los errores más grandes se produce en los valores atípicos de x (RM). Así mismo, comprobamos con las métricas los pobres resultados, aplicando regresión lineal simple, en conclusión, necesitamos más variables predictoras.

Los pobres resultados mediante LM puede deberse a la presencia de valores atípicos en la variable objetivo, lo cual pueden afectar significativamente la capacidad de tu modelo para generalizar correctamente. Si estos valores atípicos están sesgando fuertemente tus estimaciones o introduciendo ruido significativo en tus predicciones, puede ser apropiado tratarlos.

#### 1.1.1 Experimento 2: LR con tratamiento de outliers

```python
# Visualmente determinamos si hay valores atípicos:
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5)) # Crear una figura y ejes para dos gráficos
sns.boxplot(x, ax=axes[0]) # LSTAT
sns.boxplot(y, ax=axes[1])
```

Según las gráficas no se observa una cantidad grande de outliers en LSTAT, no así en MEDV, por tanto, vamos a eliminar las observaciones en las que haya outliers en esta última variable (y_column).

```python
feature = y_column

def IQRscore(df, col):
    # Determinamos los límites o bordes:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return limite_inferior, limite_superior

limite_inferior, limite_superior = IQRscore(boston, feature)

# outliers inferior:
outliers1 = boston[boston[feature] < limite_inferior][feature]

# outliers superior:
outliers2 =  boston[boston[feature] > limite_superior][feature]

#Unir los dos conjuntos:
obs_outliers = pd.concat([outliers1, outliers2], axis=0)
len(obs_outliers) # 40 observaciones afectadas por valores atípicos en MEDV
```

Como estrategia de tratamiento de valores atípicos, vamos a eliminar las 40 observaciones, esto no afectaría al tamaño total del dataset (506 observaciones).

```python
# Eliminar filas con valores atípicos en MEDV:
y_imputed = y.copy()
y_imputed.drop(obs_outliers.index, axis = 0, inplace = True)

x_imputed = x.copy()
x_imputed.drop(obs_outliers.index, axis = 0, inplace = True)
#sns.boxplot(x_imputed)
sns.boxplot(y_imputed)
```

Aunque aún quedan outliers, vamos a continuar y repetir todos los pasos previos, pero usando x_imputed y y_imputed, en estas variables se han eliminado las filas con outliers en MEDV.

```python
### Preparación de datasets (entrenamiento y test)

x_train, x_test, y_train, y_test = train_test_split(x_imputed, y_imputed, test_size=0.2, shuffle = True, random_state=10)

############################################
#PREDICCIÓN: crear el modelo predictivo.
############################################

rl1 = LinearRegression() # creamos una instancia de LinearRegression

# Instruimos al modelo que aprenda (fit) de los datos (x,y)
rl1.fit(x_train.index.values.reshape(-1, 1), y_train)  # x.reshape(-1,1): redimensiona x en arrays de 1.

# Parámetros que ha estimado la regresión lineal:
print('a = ' + str(rl1.intercept_) , 'y  b = ' + str(rl1.coef_) )

###################
#Utilizando el modelo de regresión para predecir nuevos valores (usar el método predict)
###################

# Utilizar el modelo para predecir con el conjunto de training:
y_train_predicted = rl1.predict(x_train.index.values.reshape(-1,1))

# Utilizar el modelo para predecir con el conjunto de test:
y_test_predicted = rl1.predict(x_test.index.values.reshape(-1,1))

#Ver valores reales vs predichos:
lr_predictions = pd.DataFrame({'Actual':y_test,'Predicted':y_test_predicted})
lr_predictions[:10]

###################
### Evaluación mediante métricas:
###################

# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = y_test_predicted)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))

# Coeficiente de determinación R^2 (usamos el método score).

# r^2 sobre el dataset de entrenamiento:
print('Coeficiente de Determinación R2 (train) = ' + str(r2_score(y_train, y_train_predicted)))

# r^2 sobre el dataset de test:
print('Coeficiente de Determinación R2 (test) = ' + str(r2_score(y_test, y_test_predicted)))

#a = 25.417511096931918 y  b = [-0.01406035]
```

Aunque observamos que, en general, las métricas de rendimiento mejoraron, éste no fue el caso de R^2 para el dataset de test, más bien lo contrario, el coeficiente bajó a 0,03.

Como conclusión, cuando no hay una relación lineal significativa entre la variable predictora y la variable a predecir, y/o el tamaño del dataset es muy pequeño, resulta insuficiente aplicar LR.

## 2. Predicción mediante Regresión Lineal Múltiple

Se utiliza una regresión lineal múltiple para predecir una variable objetivo Y, utilizando dos o más variables independientes X1,X2,X3...

Para aplicar este tipo de enfoque, además de lo visto, es crucial revisar los supuestos de linealidad:

1. Existen una relación lineal entre la variable dependiente y cada una de las variables independientes. El valor referencial del coeficiente de Pearson, dependerá de la cantidad de datos y de la cantidad de variables.
   
   - Como estrategia para este caso, eliminaremos todas las variables con las que MEDV tenga un coeficiente inferior a 0.4. 
<br>

2. Verificar que no existe multicolinealidad, lo cual significa que las variables independientes no están muy correlacionadas entre sí. Cuando existe multicolinealidad podemos probar:
   - Eliminando cada una de las variables X en las que exista alta correlación.
   - Aplicar el método VIF.

Todo lo anterior, es parte del feature engineering (eliminar variables, detectar las variables relevantes para el problema.

3. Por la presencia de valores atípicos podríamos probar transformando variables (por ejemplo, utilizando escalado).

3. Adicionalmente, como cuarta estrategia podemos usar modelos de ML para hacer automáticamente la selección de variables.

A continuación, vamos a ver cada enfoque.

### 2.1.  Feature engineering: borrando las variables predictoras con las que "y" tiene menor correlación

Según la matriz de correlación: y (MEDV) tiene algunas relaciones lineales con algunas variables, aunque, considerando el tamaño del dataset, no son tan significativas.

<b>¿Existe una relación lineal entre la variable dependiente y cada una de las variables independientes? </b>

Vamos a considerar un umbral de 0.4 (considerando el tamaño del dataset)

```python
# Recordemos los coeficientes de correlación
correlation['MEDV'][correlation['MEDV'].abs() <= 0.4]
```

<b>Acción</b>: Eliminar variables que tienen baja correlación (considerar un umbral, dependiendo del dataset) con la variable predictora.

```python
# Borrar las variables con  menos correlación. Vamos a tomar el rango [-0,4 a 0,4], dependiendo de la matriz de correlación este rango podría subir
X1 = X.drop(['CRIM', 'ZN', 'CHAS', 'AGE', 'DIS', 'RAD', 'B'], axis = 1)
```

```python
# Crear conjuntos de datos para el training y test
X_train, X_test, y_ent, y_test = train_test_split(X1, y, train_size=0.8, random_state=42, shuffle=True)

# Crear una instancia y entrenar el modelo con los conjuntos de datos de entrenamiento:
mlr1 = LinearRegression()
mlr1.fit(X_train, y_ent)

# Coeficientes de la función de regresión:
print('a = ' + str(mlr1.intercept_) , 'y  B = ' + str(mlr1.coef_) )

# Utilizar el modelo para generar las predicciones:
y_predict = mlr1.predict(X_test)

# EVALUACIÓN

# Verificar el R^2 (sobre el conjunto de test):
print("R2: ", r2_score(y_test, y_predict ))

# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error) con el conjunto de prueba.
mse = mean_squared_error(y_true = y_test, y_pred = y_predict)

# Errores:
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

Eliminando las variables menos correlacionadas no se obtiene un R^2 significativo, aunque subió considerablemente a 0.62.

### 2.2. Feature engineering: eliminando variables predictoras con alta correlación

Los datos no deben mostrar multicolinealidad (ocurre cuando las variables independientes están altamente correlacionadas entre sí).

Cuando existe multicolinealidad habrá problemas para determinar la variable específica que contribuye a la varianza en la variable dependiente.

Si podemos identificar plenamente, qué variables están ocasionando multicolinealidad, las podemos eliminar, sino, el mejor método para probar el supuesto es el método Variance Inflation Factor (VIF).


```python
# De las variables X que quedan, identificamos cuáles tienen alta correlación
correlation = X1.corr()

corrX = correlation[correlation.columns.difference(['MEDV'])]

sns.heatmap(corrX, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':9}, cmap='Blues')
```

Variables independientes altamente correlacionadas (>= 0.7):
*  INDUS - NOX (0.8)
*  INDUS - TAX (0.7)
*  NOX - TAX (0.7)

<b>¿Cuáles eliminamos?</b>

- Opción 1: Eliminar INDUS.
- Opción 2: Aplicar el método: Variance inflation factor (VIF) para identificar multicolinealidad

El método <b>Variance inflation factor (VIF)</b> indica el porcentaje de la varianza inflada para el coeficiente de cada variable.

- Un valor hasta de 1 indica que no hay colinealidad
- Un VIF entre 1 y 5 indica una colinealidad moderada. 
- Mientras que los valores superiores a 5 indican una alta colinealidad. 

Las matrices de correlación permiten la identificación de correlación entre pares de variables, mientras que VIF permite la evaluación general de la multicolinealidad. El VIF se puede calcular usando el paquete statsmodels.


```python
##########################
# Opción 1: eliminando las variable INDUS (y también podría ser TAX)
##########################
X2 = X1.drop(['INDUS'], axis=1)

# datasets:
x_ent, x_test, y_ent, y_test = train_test_split(X2, y, train_size=0.8, random_state=42, shuffle=True)

# Crear modelo y entrenar:
lmI = LinearRegression()
lmI.fit(x_ent, y_ent)

# Coeficientes de la función de regresión:
print('a = ' + str(lmI.intercept_) , 'y  B = ' + str(lmI.coef_) )

# Utilizar el modelo para generar las predicciones:
y_predictR = lmI.predict(x_test)

#### EVALUACIÓN:
# Verificar el R^2 (sobre el conjunto de test):

print(40*'*')
print('Coeficiente R^2: ', lmI.score(x_test, y_test)) 
print(40*'*')


# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = y_predictR)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

Al eliminar INDUS, el coeficiente R^2, pasó de 0.6209548378098455 a  0.6265791224643625, lo cual no es significativo. Por tanto, vamos a intentar eliminar la multicolinealidad, usando el segundo método (VIF ).

```python
##########################
# Opción 2: Aplicando VIF
##########################

from statsmodels.stats.outliers_influence import variance_inflation_factor

#  Calcular the VIF para todo el dataset (X)

vifF1 = pd.DataFrame()

vifF1["features"] = X1.columns
vifF1["vif_Factor"] = [variance_inflation_factor(X1, i) for i in range(X1.shape[1])]
print(vifF1)
```

En general, deberíamos apuntar a un VIF de menos de 5 (o 10) para seleccionar las variables independientes.




```python
#  Calcular the VIF para cada variable independiente:

vifF2 = pd.DataFrame()

x_DF2 = X1.drop('NOX', axis=1)
x_DF2.head()

vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

Al eliminar NOX, el VIF de algunas variables también disminuyó, aunque se mantiene alto para algunas como PTRATIO.

El proceso es recursivo, hasta dejar aquellas con VIF inferior a 10, o idealmente a 5.


```python
# Eliminar PTRATIO

x_DF2 = x_DF2.drop('PTRATIO', axis=1)

# Calcular VIF
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

```python
# Eliminar TAX
x_DF2 = x_DF2.drop('TAX', axis=1)
vifF2 = pd.DataFrame()
vifF2["features"] = x_DF2.columns
vifF2["vif_Factor"] = [variance_inflation_factor(x_DF2.values, i) for i in range(x_DF2.shape[1])]
print(vifF2)
```

Como podemos ver, después de 3 iteraciones tenemos el VIF de todas las variables restantes menos de 10. Hemos eliminado las variables con multicolinealidad y hemos identificado la lista de variables independientes que son relevantes para predecir los precios de las casas de Boston.

```python
#Creamos un nuevo modelo con las variables relevantes:

X3 = X1.drop(['NOX', 'PTRATIO', 'TAX'], axis=1) # Eliminar variables con más alto VIF.

# 80% de las observaciones quedan para el conjunto de entrenamiento y 20% para el conjunto de prueba:
x_ent, x_test, y_ent, y_test = train_test_split(X3, y, train_size=0.8, random_state=42, shuffle=True)

# Crear una instancia y entrenar el modelo:
lmI = LinearRegression()
lmI.fit(x_ent, y_ent)


# Coeficientes de la función de regresión:
print('a = ' + str(lmI.intercept_) , 'y  B = ' + str(lmI.coef_) )

# Utilizar el modelo para generar las predicciones:
y_predictR = lmI.predict(x_test)


#### EVALUACIÓN:

print(40*'*')
print('Coeficiente R^2: ', lmI.score(x_test, y_test)) # Verificar el R^2 (sobre el conjunto de test):
print(40*'*')

# Calculamos el Error Cuadrático Medio (MSE = Mean Squared Error)
mse = mean_squared_error(y_true = y_test, y_pred = y_predictR)

# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
```

En este último experimento, el rendimiento del modelo se redujo ¿Cómo lo podemos explicar? Una respuesta puede ser que no se cumple el resto de supuestos, por ejemplo, la <b>homocedasticidad</b>:

La homocedasticidad significa que la varianza de los residuos es constante. La RLM supone que la cantidad de error en los residuos es similar en cada punto del modelo lineal. Podemos graficar los residuos estandarizados contra los valores predichos para determinar si los puntos se distribuyen equitativamente entre todos los valores de las variables independientes (diagrama de dispersión).

```python
# Calcular los residuos
residuals = y_test - y_predictR

# Graficar los residuos contra los valores ajustados
plt.scatter(y_predictR, residuals)
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos vs. Valores Ajustados')
plt.axhline(y=0, color='r', linestyle='--')  # Línea horizontal en y=0 para referencia
plt.show()
```

Se comprueba que no se cumple la homocedasticidad, podría ser una razón de que no se consigue mejorar el rendimiento de RLM. Una causa de esto podría ser la presencia de valores atípicos en varias variables predictoras y en "y". Por tanto, se resalta la importancia de este paso.


### 2.3. Feature engineering: escalando datos


#### Preparar el dataset (escalando el dataset)

```python
from joblib import dump
# Escalar
scaler = preprocessing.MinMaxScaler()
X_scaled = scaler.fit_transform(X) 
X_scaled[:2]

#Guardar el modelo de escalado:
dump(scaler, 'scaler.joblib')

y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))
```

```python
print(X_scaled[0], y_scaled[0])
```

```python
# Crear datasets
#Preparación de datos:
XS_train, XS_test, ys_train, ys_test = train_test_split(X_scaled, y_scaled,test_size=0.25,random_state=10, shuffle=True)

# Crear y entrenar el modelo:
mlrs = LinearRegression() 
mlrs.fit(XS_train,ys_train) #Fit the model on the training data

#Predict price on test data
yts_predicted = mlrs.predict(XS_test)

#Actual vs predicted score
score = mlrs.score(XS_test, ys_test)

#Mean squared error
mseFull = np.mean((ys_test - mlrs.predict(XS_test))**2)

print("Accuracy of model: %f" % score)
print("Mean Squared Error: %f" % mseFull)
```

### 2.4. Feature engineering: utilizando ML para realizar el feature selection


En aprendizaje automático (ML) y estadística, la selección de características (feature selection), es el proceso de seleccionar un subconjunto de características relevantes (variables predictores) para usar en la construcción de modelos.

La premisa central cuando se utiliza una técnica de selección de características es que los datos contienen muchas características que son redundantes o irrelevantes y, por lo tanto, pueden eliminarse sin incurrir en mucha pérdida de información.

Para la selección de características vamos a usar la clase SelectFromModel. Vamos a ver cómo aplicar:

1. Por separado (primero feature selection y luego predicción), y
2. En un solo paso, mediante un pipeline.

#### Opción 1: Feature selection + training

```python
# creamos datasets, a partir del dataset original
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42, shuffle=True)

# threshold, descartará cualquier variable que caiga por debajo de este umbral
sfm = SelectFromModel(estimator=LinearRegression(), threshold=0.2).fit(X_train, y_train)
print(sfm.estimator_.coef_) # coeficientes
print(sfm.get_support()) # variables
sfm.transform(X_train)

x_new = X_train[X_train.columns[sfm.get_support()]]
print(x_new) # variables seleccionadas

mlr = LinearRegression()
mlr.fit(x_new,y_train) #Entrenar con las variables seleccionadas.

#Predecir precio, con el dataset test reducido (variables seleccionadas)
xtest_new = X_test[X_test.columns[sfm.get_support()]]
yts_predicted = mlr.predict(xtest_new)

#Actual vs predicted score
score = mlr.score(xtest_new, y_test)
#Mean squared error
mseFull = np.mean((y_test - mlr.predict(xtest_new))**2)

print("Accuracy of model: %f" % score)
print("Mean Squared Error: %f" % mseFull)
```

#### Opción 2: Definiendo un pipeline

```python
########MEDIANTE UN PIPELINE:

from sklearn.pipeline import Pipeline
pipeline1 = Pipeline([
  ('feature_selection', SelectFromModel(estimator=LinearRegression(), threshold=0.2)),
  ('classification', LinearRegression())
])
pipeline1.fit(X_train, y_train)
pipeline1.score(X_test, y_test)
```

```python
# Con el dataset escalado
from sklearn.pipeline import Pipeline
pipeline1 = Pipeline([
  ('feature_selection', SelectFromModel(estimator=LinearRegression(), threshold=0.1)),
  ('classification', LinearRegression())
])
pipeline1.fit(XS_train, ys_train)
pipeline1.score(XS_test, ys_test)
```

<b>Conclusión</b>

Según las métricas aplicadas, el dataset original (con todas las variables predictoras (Xi)) y escalado de datos, ofrece el mejor rendimiento (R^2 = 0.674559), aunque el R^2 no es un valor aceptable, más bien podríamos aplicar otras estrategias:

- Identificar/tratar valores atípicos.
- Aplicar otros métodos de predicción numérica: métodos de regresión no lineal o más robustos, como: Decision trees regression, Random Forest Regression, redes neuronales, etc.
- Luego de los experimentos, podemos comprobar que cuantas más variables tengamos mejor será el coeficiente R y, por lo tanto, no es necesario eliminar ciertas variables para mejorar nuestro modelo, al menos en lo que respecta a la regresión lineal.
